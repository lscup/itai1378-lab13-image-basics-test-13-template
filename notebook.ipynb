# Task 1: Import Libraries
In this task, we will import the necessary libraries for handling images, such as OpenCV for image processing and Matplotlib for visualizations.

```python
import cv2
import matplotlib.pyplot as plt

print("Libraries imported successfully.")
```

# Task 2: Prepare Image Source
We will not use urllib here since we are simulating this step locally.
Instead, we will prepare our sample image using pre-created files in the environment.

```python
# Utilize pre-created image in the environment
url = "sample.jpg"  # Sample image file name
file_name = "sample.jpg"
```

# Task 3: Load Image Using OpenCV
Now, we will load the image using OpenCV's `imread` function, which loads the image in BGR format.

```python
# Read the image with OpenCV
image_bgr = cv2.imread(file_name)

# Check if image is loaded
if image_bgr is not None:
    print("Image loaded successfully.")
```

# Task 4: Inspect Image Properties
We will print the type and shape of the loaded image, which gives us insight into the image's dimensions.

```python
# Print the type and shape of the image
print(f"Type: {type(image_bgr)}")
print(f"Shape: {image_bgr.shape}")
```

# Task 5: Convert BGR to RGB
Since OpenCV loads images in BGR format, we need to convert them to RGB for accurate color representation when using Matplotlib.

```python
# Convert image from BGR to RGB
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

# Task 6: Display Image
We will display the RGB image using Matplotlib, ensuring no axis ticks or labels are shown, and provide a descriptive title.

```python
# Display the RGB image
plt.imshow(image_rgb)
plt.axis('off')  # Turn off axis
plt.title('RGB Image')
plt.show()
```

# Task 7: Extract and Display Color Channels
We will extract individual Red, Green, and Blue channels from the RGB image and display each channel separately using Matplotlib.

```python
# Extract the Red, Green, Blue channels
red_channel = image_rgb[:, :, 0]
green_channel = image_rgb[:, :, 1]
blue_channel = image_rgb[:, :, 2]

# Display the channels
titles = ['Red Channel', 'Green Channel', 'Blue Channel']
channels = [red_channel, green_channel, blue_channel]

plt.figure(figsize=(12, 4))
for i, (channel, title) in enumerate(zip(channels, titles)):
    plt.subplot(1, 3, i + 1)
    plt.imshow(channel, cmap='gray')
    plt.title(title)
    plt.axis('off')
plt.show()
```
